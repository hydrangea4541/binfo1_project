# Step 4.
### Calculating Shannon Entropy

````
import collections
 
from scipy.stats import entropy
  
def estimate_shannon_entropy(dna_sequence):
    bases = collections.Counter([tmp_base for tmp_base in dna_sequence])
    # define distribution
    dist = [x/sum(bases.values()) for x in bases.values()]
 
    # use scipy to calculate entropy
    entropy_value = entropy(dist, base=2)
 
    return entropy_value
````

````
pileup_long_read_data['entropy'] = pileup_long_read_data['matches'].map(lambda x : estimate_shannon_entropy(x))
pileup_long_read_data
````

- shannon entropy 출력 후 bedtools getfasta는 어떤 기준으로..?
(entropy는 start seq +1이었는데 bedtools 는 start seq + end seq?)
- entropy만을 계산할 땐 + 1이 맞는데 getfasta 기준으론 baseread를 추가하는 게 맞아 보임. 

````
pileup_long_read_data['end'] = pileup_long_read_data['pos'] +  pileup_long_read['basereads'].apply(len)
#pileup_long_read_data['start+1'] = pileup_long_read_data['pos'] + 1
pileup_long_read_data
````

````
pileup_long_read_data_entropycut = pileup_long_read_data[pileup_long_read_data['entropy']>0.7]
````
